```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
#étape 1 : charger les ensembles des données 
data = fetch_california_housing()
```

```python
# convertir les donnée en un DataFrame (pour une mailleur visualisation)
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
print ("Apercu des données : ")
df.head()
```

```python
print ("analyse des donnée suplimentaire :")
#tracer des diagrammes pour les caracteristiques
df.hist(bins=20, figsize=(15, 10))
plt.show()
```

```python
print ("statestique descriptive pour les caracteristiques: ")
print (df.describe())
```

```python
X = data.data  #caractéristiques
Y = data.target  #cible (prix des logements en california)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
```

```python
regression_model.fit(X_train, Y_train)
```

```python
Y_pred = regression_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
#afficher les resultats
print ("model de regression linéaire entrainé avec succés.")
print(f"Erreur quadratique moyenne (RMSE) : {rmse} ")
```

```python
plt.scatter (Y_test, Y_pred, alpha=0.5)
plt.xlabel ("valeurs réelles (prix des logements)")
plt.ylabel ("prédictions")
plt.title ("regression linéaire : Valeurs Réelles vs Prédictions")
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
selector = SelectKBest (score_func=f_regression, k=5)
```

```python
X_train_selected = selector.fit_transform(X_train, Y_train)
X_test_selected = selector.transform(X_test)
```

```python
regression_model.fit(X_train_selected, Y_train)
```

```python
Y_pred_selected = regression_model.predict(X_test_selected)
```

```python
rmse_selected = np.sqrt(mean_squared_error(Y_test, Y_pred_selected))
print ("Model de regression linéaire avec caractéristiques selectionnées.")
print (f"Erreur quadratique moyenne (RMSE) avec caractéristiques selectionnées: {rmse_selected}")
```
